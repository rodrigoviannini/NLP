## removing code snippets and tags

```python
def remover(textos, regex):
    """ Função que recebe o regEX e substitui por espaço vazio """
    if type(textos) == str:
        return regex.sub('', textos)

    else: 
        return [regex.sub('', txt) for txt in textos]
```

```python
regex_html = re.compile(r'<.*?>')
questao_sem_tag = remover(questao_en, regex_html)
print(questao_sem_tag)
```

```python
def substituir_codigo(textos, regex):
    """ Função que recebe o regEX e substitui por CODE """
    if type(textos) == str:
        return regex.sub('CODE', textos)

    else: 
        return [regex.sub('CODE', txt) for txt in textos]
```

## remove double spaces and line breaks

```python
regex_espaco = re.compile(r' +')
regex_quebra_linha = re.compile(r'(\n)')

def substituir_por_espaco(textos, regex):
    """ Função que recebe o regEX e substitui por espaço """
    if type(textos) == str:
        return regex.sub(' ', textos)

    else: 
        return [regex.sub(' ', txt) for txt in textos]
```

## phrase perplexity

```python
def calcular_perplexidade(modelo, texto):
    perplexidade = 0 # para cada palavra na minhga lista de palavras vou calcular a perplexidade e acumular essa perplexidade
    palavras = WhitespaceTokenizer().tokenize(texto)
    palavras_fakechar = [list(pad_both_ends(palavra, n=2)) for palavra in palavras] 
    palavras_bigramas = [list(bigrams(palavra)) for palavra in palavras_fakechar] 

    for palavra in palavras_bigramas:
        perplexidade += modelo.perplexity(palavra)
    
    return perplexidade

   
```

P(texto) = 0
perplexidade(texto) ~= 1/P(texto)

inf = 1/0

A perplexidade é uma medida inversamente proporcional à probabilidade. Nos dados de teste, caso haja algum dado que nunca ocorreu nos dados de treino, a probabilidade para este dado é 0. Consequentemente, a perplexidade será interpretada como infinita.


Isso mesmo! A perplexidade será igual a infinito pois a probabilidade é 0.

# La Place model

Para tratar o modelo que tende ao infinito, ou seja, com a probabilidade = 0, temos que atuar na probabilidade, para isso temos algumas técnicas de suavização, como por exemplo a técnica de La Place, onde a probabilidade nunca vai ser zero

```python
from nltk.lm import Laplace

def treinar_modelo_LaPlace(lista_textos):
    todas_questoes = ' '.join(lista_textos)
    todas_palavras = WhitespaceTokenizer().tokenize(todas_questoes)
    bigrams, vocabulario = padded_everygram_pipeline(2, todas_palavras)
    modelo = Laplace(2)
    modelo.fit(bigrams, vocabulario)
    return modelo
```

modelo em ingles - frase em portugues

```python
modelo_en_LaPlace = treinar_modelo_LaPlace(en_treino)
print(calcular_perplexidade(modelo_en_LaPlace, pt_teste.iloc[0])) # primeira linha/ primeiro caso de teste
```

modelo em portugues - frase em portugues

```python
modelo_pt_LaPlace = treinar_modelo_LaPlace(pt_treino)
print(calcular_perplexidade(modelo_pt_LaPlace, pt_teste.iloc[0])) # primeira linha/ primeiro caso de teste
```

**CONCLUSÃO:** podemos inferir que a frase é escrita no idioma portugues

```python
def atribuir_idioma(lista_textos):

    idioma = []
    for texto in lista_textos:
        #calcular a perplexidade para os idiomas
        portugues = calcular_perplexidade(modelo_pt_LaPlace, texto)
        ingles = calcular_perplexidade(modelo_en_LaPlace, texto)

        #verificar qual perpexidade é menor e atribuir a lista idioma
        if ingles >= portugues:
            idioma.append('PORTUGUÊS')
        else:
            idioma.append('INGLÊS')

    return idioma
```

A principal diferença entre os métodos é que o MLE, atribui probabilidade 0 aos Ngrams que são identificados nos dados de testes, porém não constam nos dados de treinamento, o que resulta em uma perplexidade infinita. O modelo de Laplace evita este problema, utilizando uma técnica de suavização, para não zerar a probabilidade de um Ngram, mesmo que ele nunca tenha ocorrido nos dados de treinamento.


Assim, o Modelo de Laplace evita os problemas de perplexidade infinita.
